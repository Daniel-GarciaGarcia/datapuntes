```python
def square(L=1,c=(0,0)):
    square = np.array([
        [0,0],
        [0,L],
        [L,L],
        [L,0],
        [0,0]
    ])
    return (square - c - L/2).T
```

```python
q = square(5)
print(q)
plt.plot(*q,'o-')
```

```python
plt.figure(figsize=(5,5))
for i in range(1,6):
    plt.plot(*square(i),'o-')
```

```python
def square(L=1,c=(0,0),angle=0):
    # Convert degrees to radians
    angle = np.pi*angle/180

    # Rotation matrix: https://es.wikipedia.org/wiki/Matriz_de_rotaci%C3%B3n
    R = np.asarray([
        [np.cos(angle),-np.sin(angle)],
        [np.sin(angle), np.cos(angle)]
    ])
    # Points of a square of side L
    square = np.array([
        [0,0],
        [0,L],
        [L,L],
        [L,0],
        [0,0]
    ])
    return np.dot((square - c - L/2),R).T

square(angle=45)
```

```python
plt.figure(figsize=(5,5))
for i in range(1,20):
    plt.plot(*square(i, angle=5*i),'o-')
```
