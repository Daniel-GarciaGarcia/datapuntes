```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://www.emsc-csem.org/Earthquake/")
soup = BeautifulSoup(res.text)
soup
```

```python
lines = soup.select("table tbody tr")
lines[0]
```

```python
def extractData(line):
    return {
        "location": line.select_one(".tb_region").text.strip(),
        "latlong":[float(e.text.strip()) for e in line.select(".tabev1")],
        "depth": int(line.select_one(".tabev3").text),
        "magnitude": float(line.select(".tabev2")[-1].text)
    }

extractData(lines[0])
```

```python
data = [extractData(line) for line in lines]
data
```

```python
import pandas as pd

df = pd.DataFrame(data)
latlongdf = df[["latlong"]].apply(lambda x: x.latlong, result_type="expand", axis=1)
latlongdf.columns = ["lat","long"]
latlongdf
```

```python
quakedata = pd.concat([df,latlongdf], axis=1).drop(columns=["latlong"])
quakedata
```
