# GOAL: convert 2 columns into a geoJSON object

```json
{
  "type": "Point",
  "coordinates": [
    47.603122,
    -122.333253
  ]
}
```


````python
import math
def toGeoJson(longitude,latitude):
    try:
        latitude = float(latitude)
        longitude = float(longitude)
        if math.isnan(latitude) or math.isnan(longitude):
            return None
        return {
            "type":"Point",
            "coordinates":[
                longitude,
                latitude,
            ]
        }
    except Exception:
        return None
    
````

````python
toGeoJson("33","44")
````

````python
toGeoJson(33,"44")
````

````python
toGeoJson(33,"44a")
````

````python
df2["geopoint"] = df2.apply(lambda reg: toGeoJson(reg.longitude,reg.latitude) ,axis=1)
````

````python
'''
query = {
   geopoint: {
     $near: {
       $geometry: {
          type: "Point" ,
          coordinates: [ 40.759095 , -73.9871507 ]
       },
       $maxDistance: 10000,
     }
   }
}
'''
````

````python
timesSquare = [
    -73.9871507,
    40.759095
]
toGeoJson(*timesSquare)
````

````python
def nearest(center, meters=10000, geofield="geopoint"):
    return {
       geofield: {
         "$near": {
           "$geometry": toGeoJson(*center),
           "$maxDistance": meters,
         }
       }
    }
````

````python
nearest(timesSquare)
````

````python
timesSquare = [
    -73.9871507,
    40.759095
]
db.offices.find_one(nearest(timesSquare))
````

````python
import requests

def geocodeText(query):
    try:
        res = requests.get(f"https://geocode.xyz/{query}?json=1")
        print(f"Geocode url:",res.url)
        data = res.json()
        point = [data["longt"],data["latt"]]
    except:
        raise ValueError("Cannot geocode query")
    return point

geocodeText("Times Square New York")
````

````python
location = geocodeText("Puerta del Sol Madrid")
list(db.offices.find(nearest(location)))
````
