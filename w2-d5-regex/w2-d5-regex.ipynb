### Sintax
-> Special Characters:
- `.` Matches any character except a newline.
- `^` Matches the start of the string.
- `$` Matches the end of the string or just before the newline at the end of the string.
- `*` Matches 0 or more repetitions of the preceding RE.
- `+` Matches 1 or more repetitions of the preceding RE.
- `?` Matches 0 or 1 repetitions of the preceding RE.



--> Sintax:
- **Literals** `a` 
- **Alternation** `a|b`
- **Character sets** `[ab]`, `[^ab]`
- **Wildcards** `.`
- **Escape special characters** `\` (?,*,+,^,&)
- **Ranges** `[a-d]`, `[1-9]`
- **Character classes** `\w`, `\d`, `\s`, `\n`, `\W`, `\D`, `\S`
- **Quantifiers** `{2}`, `{2,}`, `{2,4}`, `?`, `*`, `+`
- **Grouping** `()`
- **Anchors** `^`, `$`

### Methods

- **sub()**
Replaces one or many matches with a string

````python
txt = "La vida es una tómbola, tom tom tómbola"
````

````python
#re.sub
#Literals
txt2 = re.sub(r't','T',txt)
txt2
````

````python
#Ranges
re.sub(r'[A-Z]','*',txt2)
````

````python
txt3 = "Weeeeeee, estoy haciendo un bootcamp de data en Ironhack???"
````

````python
#Escape special character, quantifiers
prueba = re.sub(r'\?{3}','!',txt3)
````

````python
print(prueba)
````

- **search()**
Scan through a string, looking for any location where this RE matches.

````python
#re.search
elon = "Posting Tesla Impact Report, this time with summary points & charts in my reply tweets https://tesla.com/ns_videos/2019-tesla-impact-report.pdf"
````

````python
re.search(r'Tesla',elon)
````

````python
txt = "The rain in Spain"
#\b whole words only
x = re.search(r"\bS\w+",txt)
print(x)
print(x.span())
#returns a tuple containing the start-, and end positions of the match
print(x.start())
#contains the start position of the match
print(x.end())
#contains the end position of the match
print(x.string)
#print the string passed into the function (variable 'txt')
print(x.group())
#Print the part of the string where there was a match
````

````python
txt3 = "Ese momento de tu vida en el que decides hacer el bootcamp de data"
````

````python
print(re.search(r'v\w*', txt3))
print(re.search(r'V\w*', txt3))
print(re.search(r'^E\w*', txt3))
print(re.search(r'^e\w*', txt3))
````

- **match()**
Determine if the RE matches at the **beginning** of the string.
- https://docs.python.org/2/library/re.html#search-vs-match

````python
#re.match
pattern = r"Cookie"
sequence = "I want a Cookie"
sequence2= "Cookie, I want you!"
if re.match(pattern,sequence2):
    print("Match!")
else:
    print("Nooooo")
    
````

````python
txt = "The rain in Spain"
#matches at the beginning of the string
print(re.match(r'r\w*', txt))
print(re.match(r'^r\w*', txt))
print(re.match(r'^T\w*', txt))
print(re.match(r'T\w*', txt))
````

````python
email_address = 'Please contact us at: support@data.mad.com, hola'
encuentra = re.search(r'(\w+)@([\.\w]+)', email_address)
print(encuentra)
if encuentra:
    print(encuentra.group()) # The whole matched text
    print(encuentra.group(1)) # The username (group 1)
    print(encuentra.group(2)) # The host (group 2)
````

By default in python, the ‘^’ and ‘$’ special characters (these characters match the start and end of a line, respectively) only apply to the start and end of the entire string.

Thankfully, there is a flag to modify this behavior as well. The re.MULTILINE flag tells python to make the ‘^’ and ‘$’ special characters match the start or end of any line within a string. Using this flag:

```python
match = re.search(r'(\w+)@([\w\.]+)', email_address, re.MULTILINE)
```

- **findall()**
Find all substrings where the RE matches, and returns them as a list.

````python
#re.findall
email_address = "Please contact us at: support.data@data-mad.com, amanda@ironhack.com,amanda@ironhack.com"
addresses = re.findall(r'[\w\.]+@[\w\-\.]+', email_address)
addresses

#'addresses' is a list that stores all the possible match
````

````python
print(re.findall('[^aeiou\s]',email_address))
# Match a single character not present in the list
print(re.findall('\sc\w*',email_address))
print(re.findall('^P\w*',email_address))
````

- **split()**
Returns a list where the string has been split at each match

````python
#re.split
prophet=['the{7}',
 'chosen',
 'and',
 'the\nbeloved,',
 'who',
 'was{34}',
 'a',
 'dawn',
 'unto{6}der',
 'his']
````

````python
def reference(x):
    return re.split("{\d+}",x)
````

````python
#pass the function iterating over the list
prophet_reference = list(map(reference,prophet))
print(prophet_reference)
#flatten list
pro = [e for a in prophet_reference for e in a]
print(pro)
````

### Info Sender

````python
#info_sender
info_sender = re.findall(r"From:.*",fh)
print(info_sender)
````

````python
#sender_email
emails_info['sender_email']=[]
for line in info_sender:
    res = re.findall(r"[\w\.]+@[\w\.-]+",line)
    if res:
        emails_info['sender_email'].append(res[0])
    else:
        emails_info['sender_email'].append('')
emails_info['sender_email']
````

````python
#sender name
emails_info['sender_name']=[]
for line in info_sender:
    res=re.findall(r":.*<",line)
    if res:
        emails_info['sender_name'].append(res[0][1:-1])
    else:
        emails_info['sender_name'].append('')
emails_info['sender_name']
````

## Info Dates

````python
#DATES
dates = re.findall(r"Date:.*",fh)
dates
````

````python
#email date
````

````python
emails_info['date_sent']=[]
for dat in dates:
    res = re.findall(r"\d+\s\w{3}\s\d+",dat)
    if res:
        emails_info['date_sent'].append(res[0])
    else:
        emails_info['date_sent'].append('')
emails_info['date_sent']
````

````python
emails_info['time_sent']=[]
for time in dates:
    res = re.findall(r"\d{2}:\d{2}.*\s",time)
    if res:
        emails_info['time_sent'].append(res[0][:-1])
    else:
        emails_info['time_sent'].append('')
emails_info['time_sent']
````

## Subject

````python
#subject
subjects=re.findall(r"Subject:.*",fh)
subjects
````

````python
emails_info['subject']=[]
for sub in subjects:
    res=re.findall(r":.*",sub)
    if res:
        emails_info['subject'].append(res[0][2:])
    else:
        emails_info['subject'].append('')
emails_info['subject']
````
