# Get JSON from terminal using `curl`

- get json data: `curl https://pokeapi.co/api/v2/pokemon/25`
- get json and filter with `jq`: 
    `curl https://pokeapi.co/api/v2/pokemon/25 | jq ".name,.sprites.front_default"`
    
    
Refs 
- https://stedolan.github.io/jq/tutorial/
