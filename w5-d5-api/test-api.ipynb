```python
import requests

def getInstructor():
    res = requests.get("http://localhost:3000")
    data = res.json()
    return data["instructor"]

def translateInstructor(ins):
    translation = {
        "Amanda":0,
        "Marc":1,
        "Felipe":2
    }
    return translation[ins] if ins in translation else -1
```

```python
instructors = []
for i in range(200):
    x = getInstructor()
    instructors.append(translateInstructor(x))
```
