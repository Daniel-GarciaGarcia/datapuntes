```python
import requests


def getPokemon(pokeId):
    print(f"Requesting pokemon {pokeId}")
    res = requests.get(f"https://pokeapi.co/api/v2/pokemon/{pokeId}")
    data = res.json()
    return {
        "pokeId":pokeId,
        "name":data["name"],
        "image":data["sprites"]["front_default"],
        "moves": [move["move"]["name"] for move in data["moves"]]
    }
    
```

```python
pokemons = [getPokemon(i) for i in range(1,152)]
```

```python
import pandas as pd

pokeDf = pd.DataFrame(pokemons)
```

```python
pokeDf.to_json("data/pokemons_pokeapi.json", orient="records")
```
