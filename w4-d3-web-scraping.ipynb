```python
import numpy as np
import pandas as pd

data_raw = [tag.text for tag in medallero_soup.find_all("td")]
m = np.asarray(data_raw)
print(data_raw)
q = m.reshape((int(m.size/6),6))
print(q)
pd.DataFrame(q)
```

```python
def procesaFila(fila):
    celdas = fila.find_all("td")
    return {
        "pais":celdas[1].find("a").text.strip(),
        "oro":int(celdas[2].text),
        "plata":int(celdas[3].text),
        "bronce":int(celdas[4].text)
    }

fila = medallero_soup.find_all("tr")[1]
print(fila)
procesaFila(fila)
```

```python
data = [procesaFila(tag) for tag in medallero_soup.find_all("tr")[1:]]
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df["total"] = df[["oro","plata","bronce"]].sum(axis=1)
```

```python
url = "https://deliveroo.es/es/restaurants/madrid/latina"
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
import re
card = soup.select(".HomeFeedUICard-ebc18e264cc4b96c")[0]

def processCard(card):
    raw_data = [tag.text for tag in card.select(".ccl-19882374e640f487")]
    test = re.match(r"\d+\.\d+",raw_data[4])
    if test:
        grade = test.group()
    else:    
        #raise ValueError("Cannot parse data")
        return None
    return {
        "name":raw_data[3],
        "grade":float(grade)
    }
processCard(card)
```

```python
restaurants = [processCard(card) for card in soup.select(".HomeFeedUICard-ebc18e264cc4b96c")]

restaurants = list(filter(lambda e: e, restaurants))
```

```python
df = pd.DataFrame(restaurants)
df.sort_values(by="grade", ascending=False)
```
